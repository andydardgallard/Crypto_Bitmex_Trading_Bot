```python
import requests
import json
import csv

def write_csv(filename,new_line):
    with open(filename, "a" ) as fout:
        writer=csv.writer(fout)
        writer.writerow((new_line))
```

```python
symbol = "XBTZ17"
stop_url="https://www.bitmex.com/api/v1/quote/bucketed?binSize=5m&partial=false&symbol="+symbol+"&count=1&reverse=true"
stop_html=requests.get(stop_url).text
stop_data=json.loads(stop_html)

for j in stop_data:
    stop=j['timestamp']
```

```python
base_url="https://www.bitmex.com/api/v1/quote/bucketed?binSize=5m&partial=false&symbol="+symbol+"&count=500&reverse=false"
```

```python
html=requests.get(base_url).text
data=json.loads(html)

filename_BID=data[0]["symbol"]+"_BID.txt"
filename_ASK=data[0]["symbol"]+"_ASK.txt"


while data[-1]["timestamp"] != stop:    
    
    for i in data:
        YYYY=str(i["timestamp"])[:4]
        MM=str(i["timestamp"])[5:7]
        DD=str(i["timestamp"])[8:10]
        DATE=YYYY+MM+DD
    
        HH=str(i["timestamp"])[11:13]
        Min=str(i["timestamp"])[14:16]
        Sec="00"
        TIME=HH+Min+Sec
    
        BID=str(i["bidPrice"])    
        BID_V=str(i["bidSize"])       

        new_line=[DATE, TIME, BID, BID_V]
        if BID !="None":    
            write_csv(filename_BID, new_line)
            
        ASK=str(i["askPrice"])    
        ASK_V=str(i["askSize"])     

        new_line=[DATE, TIME, ASK, ASK_V]
        if ASK !="None":    
            write_csv(filename_ASK, new_line)            
        
            
    Min=int(Min)+5

    if Min>55:
        Min="00"
        if len(HH)==1:
            HH="0"+str(int(HH)+1)
    
    if Min==5:       
        url=base_url+"&startTime="+i["timestamp"][:11]+HH+"%3A"+"0"+str(Min)+"%3A00.000Z"        
    else:        
        url=base_url+"&startTime="+i["timestamp"][:11]+HH+"%3A"+str(Min)+"%3A00.000Z"
        
    html=requests.get(url).text
    data=json.loads(html)
    
    print url  
    
for i in data:
    YYYY=str(i["timestamp"])[:4]
    MM=str(i["timestamp"])[5:7]
    DD=str(i["timestamp"])[8:10]
    DATE=YYYY+MM+DD
    
    HH=str(i["timestamp"])[11:13]
    Min=str(i["timestamp"])[14:16]
    Sec="00"
    TIME=HH+Min+Sec
    
    BID=str(i["bidPrice"])    
    BID_V=str(i["bidSize"])       

    new_line=[DATE, TIME, BID, BID_V]
    if BID !="None":    
        write_csv(filename_BID, new_line)
            
    ASK=str(i["askPrice"])    
    ASK_V=str(i["askSize"])     

    new_line=[DATE, TIME, ASK, ASK_V]
    if ASK !="None":    
        write_csv(filename_ASK, new_line)
```
